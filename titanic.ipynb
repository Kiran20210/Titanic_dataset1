```python
def comp_type(row):
    if row['psg_cnt']==1:
        return 'solo'
    elif row['family_count']==0 & row['psg_cnt']>1:
        return 'friends'
    else:
        return'family'

df['companion_type']=df.apply(comp_type,axis=1)
```

```python
fig,a=plt.subplots(nrows=2,figsize=(10,6))
sns.kdeplot(df,x='age',ax=a[0])
sns.kdeplot(df,x='total_fare',ax=a[1])
a[0].set_xticks(range(0,85,3))
a[1].set_xticks(range(0,500,40))
a[0].grid()
a[1].grid()
plt.tight_layout()
plt.show()
```
